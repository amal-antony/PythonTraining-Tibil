## 3. Stacked Histograms
The stacked bar chart (aka stacked bar graph) extends the standard bar chart from looking at numeric values across one categorical variable to two.<br>
Each bar in a standard bar chart is divided into a number of sub-bars stacked end to end, each one corresponding to a level of the second categorical variable.<br>
Stacked Chart. Bar charts are best used when showing comparisons between categories. <br>
The bars are proportional to the values they represent and can be plotted either horizontally or vertically. One axis of the chart shows the specific categories being compared, and the other axis represents discrete values.

### Adding legend and label

```python
genre_list=list()
labels=list()
for gen in movies.Genre.cat.categories:
    genre_list.append(movies[movies.Genre==gen].BudgetInMillions)
    labels.append(gen)
print(genre_list)
```

```python
hist=plt.hist(genre_list,bins=30,stacked=True,rwidth=1,label=labels)
plt.legend()
plt.show()
```

## 5. SubPlots
functions that enable us to create more sophisticated visualizations to combine visualizations Together<br>

### f,axes=plt.subplots(no _of_rows,no_of_columns)
Creates subplots of no of rows each having no of columns subplots<br>
2,4 means creates 2 rows each having 4 subplots<br>
f stands for figure , it can be fig or any same , it is actually figure object of figure class in matplotlib

```python
f,axes=plt.subplots(4,2)
```

### Placing charts
Additional parameter called ax can be used to place each chart we need in whicever of the available subplots we created.<br>
Assign the value to 
### ax  
array of axes if more than 1 subplot is created<br> 
as axes[k] <br>
where axes stands for 2 nd parameter of f,axes and k is the index we want to place

```python
f,axes=plt.subplots(1,2,figsize=(12,6))
k1=sns.kdeplot(x=movies.BudgetInMillions,y=movies.AudienceRatings,ax=axes[0])
k2=sns.kdeplot(x=movies.BudgetInMillions,y=movies.CriticRatings,ax=axes[1])
```

# Setting Limits for values in x and y axis
xlim()-pass limts for x axis<br>
ylim()-pass limits for y axis <br>
This method is called on the variable that was used to create the graph<br>

#### sharex=True,
#### sharey=True
Will share x and y values for all subplots

```python
f,axes=plt.subplots(1,2,figsize=(12,6))
k1=sns.kdeplot(x=movies.BudgetInMillions,y=movies.AudienceRatings,ax=axes[0])
k2=sns.kdeplot(x=movies.BudgetInMillions,y=movies.CriticRatings,ax=axes[1])
k1.set(xlim=(-20,160))
```

```python
f,axes=plt.subplots(1,2,figsize=(20,6),sharex=True)
k1=sns.kdeplot(x=movies.BudgetInMillions,y=movies.AudienceRatings,ax=axes[0])
k2=sns.kdeplot(x=movies.BudgetInMillions,y=movies.CriticRatings,ax=axes[1])
k1.set(xlim=(-20,140))
```
