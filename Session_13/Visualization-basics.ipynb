## Adding a Legend
A box that shows different marker symbols on the left and data it reoresents on the right.<br>
Legend is used after adding label
### plt.legend()<br>

#### 1.loc
A property that tells where the box must be shown <br>
Has the following options<br>
a.upper left<br>
b.upper right<br>
c.lower left<br>
d.lower left<br>

#### 2. bbox_to_anchor

1.Takes a tuple with 2 values<br>
2.Each value specifies where we want to place our legend w.r.t to x and y axis <br>
3.Eg: (1,1) means 1 unit to the right of x axis and 1 unit above y axis , where the entire unit of graph is considered as a rectangle with 1 unit length and breadth





```python
plt.plot(Salary[0],c='Black',ls='--',marker='s',ms=7,label=Players[0])
plt.plot(Salary[1],c='Blue',ls='--',marker='s',ms=7,label=Players[1])
plt.plot(Salary[2],c='Red',ls='--',marker='s',ms=7,label=Players[2])
plt.plot(Salary[3],c='Green',ls='--',marker='s',ms=7,label=Players[3])
plt.plot(Salary[4],c='Yellow',ls='--',marker='s',ms=7,label=Players[4])
plt.plot(Salary[5],c='Orange',ls='--',marker='s',ms=7,label=Players[5])
plt.plot(Salary[6],c='Brown',ls='--',marker='s',ms=7,label=Players[6])
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
plt.show()
```

### Creating the same plot as a function

```python
def myplot(playerlist=Players,data_matrix=Salary):
    colors={"KobeBryant":'Black',"JoeJohnson":'Blue',"LeBronJames":'Red',"CarmeloAnthony":'Green',"DwightHoward":'Yellow',"ChrisBosh":'Orange',"ChrisPaul":'Brown',"KevinDurant":'Grey',"DerrickRose":'Magenta',"DwayneWade":'cyan'}
    markers={"KobeBryant":'o',"JoeJohnson":'v',"LeBronJames":'^',"CarmeloAnthony":'<',"DwightHoward":'>',"ChrisBosh":'h',"ChrisPaul":'d',"KevinDurant":'x',"DerrickRose":'s',"DwayneWade":'p'}
    line_styles={"KobeBryant":'-',"JoeJohnson":'--',"LeBronJames":'-.',"CarmeloAnthony":':',"DwightHoward":'solid',"ChrisBosh":'',"ChrisPaul":'dashdot',"KevinDurant":'dotted',"DerrickRose":'dashed',"DwayneWade":'None'}
    for name in playerlist:
        plt.plot(data_matrix[Pdict[name]],c=colors[name],ls=line_styles[name],marker=markers[name],ms=7,label=name)
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.show()
```
