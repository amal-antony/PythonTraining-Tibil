### In memory computations- 
All data processed in harddisk in data nodes - data is read from harddisk , it goes to RAM , operations performed on it and goes back to harddisk<br>
Entire data is read from harddisk not just portion that is required , so entire data is stored in RAM which makes computing easier<br>
Data remains in RAM and only after entire operation is completed it is stored into harddisk<br>
Decreases latency<br>

This is not the case with Hadoop , It reads only requird data from harddisk and after operation store it back to harddisk . Every time before and after processing the data it is stored in harddisk whch causes latency<br>


### Lazy Excecution 
Whenever we apply some function to read the data ,it will not read the data until and unless we perform some operations on the data<br>
Eg: pd.read_csv() it just store data to RAM , in case of Spark it dosnot perform like that it doesnt store data inside the RAM until we perfrom any operation on it

## Comparison b/w Mapreduce and Spark
### Mapreduce
1.Computing framwwork , Open source<br>
2.Faster than traditional system doesnt leverage memory of hadoop cluster to maximum<br>
3.Disk oriented and higher latency<br>
4.Cheaper option available <br>
5.Java Based and thus MApReduce pipelines is costly and complex and lengthy as its purely Java<br>
6.Batch Processing<br>
7.Fault Tolerance , Highly Scalable and Cross Platform<br>
8.No inbuilt Support to SQL,ML,RT<br>
9.Basic data Processing engine<br>
10.Runs well on commodity software<br>
11.Tested on and can run in 15000 nodes<br>
12.Small Dataset


### Spark
1.Computing framwwork , Open source<br>
2.Faster than MApreduce batch processing 10-100 times faster uses full capacity of Hadoop cluster<br>
3.Lower latency , caching partial results across memory of distributed software . Stores data in memory<br>
4.Requires a lot of RAM , increase in cluster and cost ,reduce no of reading cycles <br>
5.MApReduce pipelines is costly and complex and lengthy as its purely Java<br>
6.Batch/Iterative/Real Time/Interactive Processing<br>
7.Fault Tolerance , Highly Scalable and Cross Platform<br>
8.Has Inbuilt Support to SQL,ML,RT<br>
9.Data Analytics Engine . Choice for data scientist<br>
10.Needs mid to highlevel hardware<br>
11.Tested on and can run in 8000 nodes<br>
12.Large Dataset and complex computations
